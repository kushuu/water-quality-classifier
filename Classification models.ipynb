```python
def work_amm(val):
    if(val != "#NUM!"):
        return float(val)
    else:
        return np.nan

def work_issafe(val):
    if(val != "#NUM!"):
        return float(val)
    else:
        return np.nan
    
df.ammonia = df.ammonia.apply(work_amm)
df.is_safe = df.is_safe.apply(work_issafe)
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
X = df.drop('is_safe',1)
y = df['is_safe']

features = X.columns
```

## Since our dataset in clean, working with the classification model now

```python
from sklearn.feature_selection import SelectKBest, f_classif, chi2
```

```python
fs = SelectKBest(score_func=f_classif, k=8)
```

```python
z = fs.fit_transform(X, y)
```

```python
supported_features = fs.get_support()
```

```python
selected_features = features[supported_features]
```

## Here are the selected features from the pool of 20 features.

```python
selected_features
```

```python
X = X[selected_features]
```

```python
# importing the models to be used for training.

from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
# will be defining a class for ensemble learning of all the classification models.

from statistics import mode

class my_classifier():
    def __init__(self, classifiers):
        self._classifiers = classifiers

    def classify(self, input_params):
        votes = []
        for c in self._classifiers:
            v = c.predict(input_params.reshape(1,-1))[0]
            # print(str(c)[:-2], "predicted", v)
            votes.append(v)
        return mode(votes)

    def confidence(self, input_params):
        votes = []
        for c in self._classifiers:
            v = c.predict(input_params.reshape(1,-1))[0]
            votes.append(v)

        choice_votes = votes.count(mode(votes))
        conf = choice_votes / len(votes)
        return conf
```

```python
# splitting the data for training and testing.

from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, train_size=0.3, random_state=100)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
models = []

for model in [LogisticRegression(), SGDClassifier(), RandomForestClassifier(), SVC(), KNeighborsClassifier(), GaussianNB(), ExtraTreesClassifier()]:
    model_obj = model

    begin_time = datetime.now()
    model_obj.fit(X_train, y_train)
    end_time = datetime.now()
    train_time = end_time-begin_time

    test_time_begin = datetime.now()
    acc = model_obj.score(X_test, y_test)
    test_time_end = datetime.now()
    test_time = test_time_end-test_time_begin

    models.append(
        [
            model_obj, str(train_time.seconds) + '.' + str(train_time.microseconds), str(test_time.seconds) + '.' + str(test_time.microseconds), acc
        ]
    )
    save_model = open(os.path.join(os.getcwd(), "pickled models/" + str(model)[:-2] + "_model.pickle"),"wb")
    pickle.dump(model_obj, save_model)
    save_model.close()
```

```python
for model in models:
    print(model)
```

```python
# sorting the models in the list according to their accuracies.
models.sort(key = lambda x : x[-1], reverse=True)
```

```python
print("Model name\t\t\tTraining time\tTesting time\tAccuracy","="*80, sep='\n')

s = [[str(e) for e in row] for row in models]
lens = [max(map(len, col)) for col in zip(*s)]
fmt = '\t'.join('{{:{}}}'.format(x) for x in lens)
table = [fmt.format(*row) for row in s]
print('\n'.join(table))
```

## Testing what is built till now.

```python
classifier = my_classifier([model[0] for model in models])
```
